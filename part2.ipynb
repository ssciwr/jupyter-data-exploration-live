## Selecting rows/columns: `iloc` 

```python
# row with index label "0"
df.loc[0]
```

```python
# equivalent
df.loc[0, :]
```

```python
# Age column
df.loc[:, "Age"]
```

```python
# Slices are *inclusive*
df.loc[:, "Age":"Eye colour"]
```
