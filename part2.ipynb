## Plotting

```python
# import matplotlib
import matplotlib.pyplot as plt
```

```python
# scatter graph of two columns
df.plot.scatter(x="Age", y="Height")
```

```python
fig, axs = plt.subplots(figsize=(20,12))
df.plot.scatter(x="Age", y="Height", ax=axs, label="Data")
plt.title("Height vs Age")
plt.show()
```

```python
fig, axs = plt.subplots(figsize=(12,12))
for sex, colour in [("Male", "black"), ("Female", "blue")]:
    df.loc[df.Sex == sex].plot.scatter(x="Age", y="Height", ax=axs, label=sex, color=colour)
axs.set_title("Height vs Age")
# plt.title("Height vs Age")
plt.show()
```

```python
df["Eye colour"].value_counts()
```

```python
fig, axs = plt.subplots(ncols=3)
df["Eye colour"].value_counts().plot.pie(ax=axs[0])
df["Wears glasses"].value_counts().plot.pie(ax=axs[1])
df["Sex"].value_counts().plot.pie(ax=axs[2])
```
