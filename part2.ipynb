## Selecting rows/columns: `iloc`

```python
# first column
df.iloc[:, 0]
```

## Selecting rows/columns: `iloc` 

```python
# row with index label "0"
df.loc[0]
```

```python
# equivalent
df.loc[0, :]
```

```python
# Age column
df.loc[:, "Age"]
```

```python
# Slices are *inclusive*
df.loc[:, "Age":"Eye colour"]
```

```python
# important difference!
df.iloc[0:2, :]
```

```python
df.loc[0:2, :]
```

```python
df.loc[[2,3,4], ["Eye colour", "Age"]]
```
