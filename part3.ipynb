```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
url = "https://data.london.gov.uk/download/animal-rescue-incidents-attended-by-lfb/f43b485e-fb35-419c-aa7a-fa75676e5835/Animal%20Rescue%20incidents%20attended%20by%20LFB%20from%20Jan%202009.csv"
```

```python
df = pd.read_csv(url, encoding="unicode_escape")
```

```python
df.columns
```

```python
pd.to_datetime(df["DateTimeOfCall"], dayfirst=True)
```

```python
df["DateTimeOfCall"] = pd.to_datetime(df["DateTimeOfCall"], dayfirst=True)
```

```python
df["DateTimeOfCall"]
```

```python
df.set_index("DateTimeOfCall", inplace=True)
```

```python
df
```

```python
# slice by datetime
df.loc["2009.12.01":"2009.12.07", :]
```

```python
# resample
df.resample("Y")["CalYear"].count().plot()
```

```python
fig, axs = plt.subplots(figsize=(16, 4), ncols=2)
df.resample("Y")["IncidentNotionalCost(£)"].sum().plot(
    title="Year total cost", ax=axs[0]
)
df.resample("Y")["HourlyNotionalCost(£)"].mean().plot(
    title="Average hourly cost", ax=axs[1]
)
plt.show()
```
